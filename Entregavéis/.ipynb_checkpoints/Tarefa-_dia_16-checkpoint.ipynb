```python
import pandas as pd

# Caminho completo do arquivo
csv_path = r"C:\Users\Visitante\Desktop\Bootcamp\Data-Engineering-and-Big-Data-Project\Datasets\worldometer_data.csv"

# Ler o CSV para um DataFrame
df = pd.read_csv(csv_path)

# Mostrar as 5 primeiras linhas
df.head()

```

```python
df = df.drop(columns=['NewCases'])
df = df.drop(columns=['NewDeaths'])
df = df.drop(columns=['NewRecovered'])
```

```python
df.head()
```

```python
from sqlalchemy import create_engine
import urllib

params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=(localdb)\MSSQLLocalDB;"   # ðŸ‘ˆ seu servidor local
    "DATABASE=Bootcamp;"            # ðŸ‘ˆ nome do banco que vocÃª criou
    "Trusted_Connection=yes;"         # ðŸ‘ˆ autenticaÃ§Ã£o do Windows
    "Encrypt=no;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
with engine.connect() as conn:
    print("âœ… ConexÃ£o bem-sucedida!")
```

```python
df.to_sql(
    'covid',         # nome da tabela no SQL
    con=engine,      # conexÃ£o que vocÃª criou
    if_exists='replace',  # substitui se jÃ¡ existir
    index=False      # nÃ£o envia o Ã­ndice do pandas
)
print("âœ… Tabela enviada para o SQL Server com sucesso!")
```
